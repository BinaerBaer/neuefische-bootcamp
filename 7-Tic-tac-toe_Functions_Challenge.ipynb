````python
class TTTBoard(object):
    
    def __init__(self):
        self.board = []
        self.buildBoard()
        self.playerSymb = ['O', 'X']
    
    def buildBoard(self):
        self.board = [[(x, y) for y in range(3)]
                          for x in range(3)]
    
    # überschreibt print()-Methode
    def __str__(self):
        string = ""
        for line in self.board:
            string += str(line) + "\n"
        return string
    
    def start_game(self):
        self.buildBoard()
        print(self)
        self._turn = 0
        
        while (self._turn < 9) and (not self._checkWin()):
            playerSymb = self.playerSymb[self._turn % 2]
            
             
            # BREAK Funktion außer Kraft 
            # brk = input('Hand keyboard over to player "' + playerSymb + '" and press "Enter" or "q" to quit: ')
            brk = 1
            if not brk == 'q':
                self._askPlayer(playerSymb)
            else:
                print('Game quit, Player', playerSymb, 'loses')
                break
            print(self)
            self._turn += 1
        
        if (self._turn == 9) and (not self._checkWin()):
            print('We are out of tiles, it\'s a tie!')
            
        return None
            
    def _askPlayer(self, ID):
        valid_input = False
        
        while not valid_input:
            tile_row_str = input('Player "' + ID + '" Please enter tile row (0/1/2): ')
            tile_col_str = input('Player "' + ID + '" Please enter tile column (0/1/2): ')
            if (tile_row_str in ['0', '1', '2']) and (tile_col_str in ['0', '1', '2']):
                tile_row = int(tile_row_str)
                tile_col = int(tile_col_str)
            else:
                print('I am sorry, something went wrong, please enter valid row and column number')
            
            if type(self.board[tile_row][tile_col]) != tuple:
                print('I am sorry, tile is already taken, please enter valid row and column number')
            else:
                valid_input = True
                self.board[tile_row][tile_col] = '  ' + ID + ' '
                
    def _checkWin(self):
        for ID in self.playerSymb:
            # create check lists for player <ID>
            rows = [all([ID in self.board[row][col] for row in range(3)]) 
                    for col in range(3)]            
            cols = [all([ID in self.board[row][col] for col in range(3)]) 
                    for row in range(3)]            
            diag1 = [all([ID in self.board[i][i] for i in range(3)])]            
            diag2 = [all([ID in self.board[2-i][i] for i in range(3)])]
            
            #any Trues for player <ID>?
            if any(rows + cols + diag1 + diag2):
                print('Player "', ID, '" wins! Congratulations')
                return True
````
